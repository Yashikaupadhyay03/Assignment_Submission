```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Prepare the data
X = data.drop('letter', axis=1)  # Features
y = data['letter']  # Target variable
```

```python
# Encode target variable
y_encoded = pd.get_dummies(y).values
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# 2. Model Building

```python
# Build the ANN model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden layer
    layers.Dense(y_encoded.shape[1], activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {accuracy:.2f}')
```

# 3. Hyperparameter Tuning

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
# Define hyperparameters for tuning
hidden_layers = [(10,), (20,), (10, 10), (20, 20)]  # Different configurations of hidden layers
activation_functions = ['relu', 'tanh']  # Activation functions
learning_rates = [0.001, 0.01, 0.1]  # Learning rates

# Store results
results = []

# Grid search for hyperparameter tuning
for layers in hidden_layers:
    for activation in activation_functions:
        for lr in learning_rates:
            # Create and train the model
            model = MLPClassifier(hidden_layer_sizes=layers, activation=activation, learning_rate_init=lr, max_iter=1000, random_state=42)
            model.fit(X_train, y_train)
            # Predict and evaluate the model
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            results.append({'hidden_layers': layers, 'activation': activation, 'learning_rate': lr, 'accuracy': accuracy})

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Show the results
results_df.sort_values(by='accuracy', ascending=False).reset_index(drop=True)
```

# 4. Evaluation

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Train a model with default hyperparameters
default_model = MLPClassifier(max_iter=1000, random_state=42)
default_model.fit(X_train, y_train)

# Predict using the default model
y_pred_default = default_model.predict(X_test)

# Calculate metrics for the default model
accuracy_default = accuracy_score(y_test, y_pred_default)
precision_default = precision_score(y_test, y_pred_default, average='weighted')
recall_default = recall_score(y_test, y_pred_default, average='weighted')
f1_default = f1_score(y_test, y_pred_default, average='weighted')

# Calculate metrics for the tuned model
best_tuned_model = MLPClassifier(hidden_layer_sizes=(20, 20), activation='tanh', learning_rate_init=0.001, max_iter=1000, random_state=42)
best_tuned_model.fit(X_train, y_train)

y_pred_tuned = best_tuned_model.predict(X_test)

accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned, average='weighted')
recall_tuned = recall_score(y_test, y_pred_tuned, average='weighted')
f1_tuned = f1_score(y_test, y_pred_tuned, average='weighted')

# Create a summary of the results
performance_summary = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Default Model': [accuracy_default, precision_default, recall_default, f1_default],
    'Tuned Model': [accuracy_tuned, precision_tuned, recall_tuned, f1_tuned]
}

performance_summary_df = pd.DataFrame(performance_summary)
performance_summary_df
```

**The default model slightly outperforms the tuned model across all metrics.**
