```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/glass.xlsx', sheet_name='glass')
data.head(50)
```

# EDA

```python
plt.figure(figsize=(15, 7))
plt.pie(data['Type'].value_counts(), autopct='%1.1f%%', labels=data['Type'].unique())
plt.title('Pie Chart of type Distribution', fontsize=15)
plt.show()
```

**Type: Type of glass: (class attribute)**

1 --> building_windows_float_processed

```python
plt.figure(figsize = (10, 15))
for i, col in enumerate(data.columns[:-1], 1):
    plt.subplot(5, 2, i)
    sns.histplot(x = data[col], hue = data["Type"], multiple = "dodge")
    plt.title(f"Distribution of {col} Data")
    plt.tight_layout()
    plt.xticks(rotation = 90)
    plt.plot()
```

THE **RI** OF **TYPE 1 AND TYPE 2** GLASS IS BETWEEN **1.515 TO 1.525.**

THE **Na** OF **TYPE 1 AND TYPE 2** IS BETWEEN **11 TO 14** AND OF **TYPE 5, 6, 7** IS **ABOVE 13**.

THE **Fe** OF **TYPE 5, 6 AND 7** IS **NOT ABOVE 0.1** AND OTHER TYPES HAVE BETWEEN **0.1 TO 0.3**

THE **Ba** of **TYPE 1, 2 AND 3** IS **0**.

THE **K** IS **NOT MORE THAN 1** IN ANY TYPE.



```python
element_concentrations = data.groupby('Type')[['Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']].mean()

plt.figure(figsize=(12, 6))
element_concentrations.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Glass Type')
plt.ylabel('Average Concentration')
plt.title('Element Concentration by Glass Type')
_ = plt.legend(loc='upper right')
```

**EVERY GLASS TYPE HAS HIGHER CONCENTRATION OF SI (SILICON), I.E IT IS VERY IMPORTANT VARIABLE.**

**NA (SODIUM) IS ALSO PRESENT IN EQUAL AMOUNT IN EACH TYPE.**

**K (POTASSIUM) IS ONLY PRESENT IN TYPE 5.**

**BA (BARIUM) IS ONLY PRESENT IN ONLY TYPE 7.**

```python
y = data['Type']
x = data.drop('Type', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns)

x_scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

# BAGGING

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators=10, random_state=42,max_samples=0.4,max_features=0.6,max_depth=6)
RFC.fit(x_train, y_train)
print("Bagging-Training score:",RFC.score(x_train, y_train))
print("Bagging-Test score:",RFC.score(x_test, y_test))
```

**BOOSTING**

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(n_estimators=150, random_state=42, learning_rate=0.1)
GBC.fit(x_train, y_train)
print("Boosting-Training score:",GBC.score(x_train, y_train))
print("Boosting-Test score:",GBC.score(x_test, y_test))
```

# CONCLUSION

**BAGGING :** Model is performing good on both training and testing data.
