# **K-MEANS CLUSTERING**

# Without outliers treatment

```python
from sklearn.cluster import KMeans

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(final_data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(final_data)
```

```python
final_data['Kmeans-Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(final_data['Balance'], final_data['Flight_miles_12mo'], c=final_data['Kmeans-Cluster'], cmap='viridis')
plt.title('K-Means Clustering of East-West Airlines Customers')
plt.xlabel('Balance')
plt.ylabel('Flight Miles in 12 Months')
plt.colorbar(label='Cluster')
plt.show()

print('Clustering and visualization done.')
```

# After outliers treatment

```python
#LOAD THE DATASET
df = pd.read_excel('/content/EastWestAirlines.xlsx', sheet_name= 'data')

# Function to remove outliers using IQR method
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Select numerical columns for outlier treatment
numerical_columns = ['Balance', 'Qual_miles', 'cc1_miles', 'cc2_miles', 'cc3_miles',
                     'Bonus_miles', 'Bonus_trans', 'Flight_miles_12mo', 'Flight_trans_12',
                     'Days_since_enroll']

# Remove outliers
data_cleaned = remove_outliers(data, numerical_columns)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_cleaned)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Add the cluster labels to the original dataframe
data_cleaned['Cluster'] = clusters

# Visualize the clustering result using a scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(data_cleaned['Balance'], data_cleaned['Flight_miles_12mo'],
                      c=data_cleaned['Cluster'], cmap='viridis', alpha=0.6)
plt.title('K-Means Clustering of East-West Airlines Customers (After Outlier Treatment)')
plt.xlabel('Balance')
plt.ylabel('Flight Miles in 12 Months')
plt.colorbar(scatter, label='Cluster')
plt.show()

```

```python
from sklearn.metrics import silhouette_score

# Silhouette score for clustering without outlier treatment
silhouette_before = silhouette_score(final_data.drop('Kmeans-Cluster', axis=1), y_kmeans)
print("Silhouette Score (Before Outlier Treatment):", silhouette_before)

# Silhouette score for clustering after outlier treatment
silhouette_after = silhouette_score(data_scaled, clusters)
print("Silhouette Score (After Outlier Treatment):", silhouette_after)
```

**Insights:**

* **The Silhouette Score for the original data is higher (0.3304) compared to the cleaned data (0.2980), indicating that the clustering structure is better in the original dataset.**

# **AGGLOMERATIVE CLUSTERING**

```python
# Import necessary libraries for Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns

# Perform Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clusters = agg_clustering.fit_predict(final_data)

# Add the cluster labels to the original dataframe
final_data['Agglomerative_Cluster'] = agg_clusters

# Visualize the clustering result using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=final_data['Balance'], y=final_data['Flight_miles_12mo'], hue=final_data['Agglomerative_Cluster'], palette='viridis', s=100)
plt.title('Agglomerative Clustering of East-West Airlines Customers')
plt.xlabel('Balance')
plt.ylabel('Flight Miles in 12 Months')
plt.legend(title='Cluster')
plt.show()

print('Agglomerative clustering and visualization done.')
```

**We can observe three distinct clusters:**

**Cluster 0 (blue):** This cluster seems to contain customers with lower balances and fewer flight miles. These might be occasional travelers or new customers.

**Cluster 1 (orange):** This cluster appears to have customers with moderate to high balances and a wide range of flight miles. These could be regular customers with varying travel frequencies.

**Cluster 2 (green):** This cluster contains customers with very high balances and a wide range of flight miles. These might be the airline's most valuable customers, possibly including business travelers or very frequent flyers.


```python
# Silhouette score for clustering after outlier treatment
silhouette_agg = silhouette_score(final_data, agg_clusters)
print("Silhouette Score og Agglomerative Clustering:", silhouette_agg)
```

**Agglomerative clustering is performing slightly better based on the Silhouette Score.**

# **DB SCAN CLUSTERING**

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
# Select numerical columns for DBSCAN
numerical_columns = ['Balance', 'Flight_miles_12mo']
X = final_data[numerical_columns].values
```

```python
# Compute the k-distance graph
neigh = NearestNeighbors(n_neighbors=2)
nbrs_distances, nbrs_indices = neigh.fit(X).kneighbors(X)
distances = nbrs_distances[:, 1]
distances = np.sort(distances, axis=0)

# Plot k-distance graph
plt.figure(figsize=(10, 5))
plt.plot(distances)
plt.title('K-distance Graph')
plt.xlabel('Points sorted by distance')
plt.ylabel('Epsilon')
plt.show()
```

```python
# Choose epsilon based on the k-distance graph
epsilon = 0.5  # Adjust this value based on the graph

# Perform DBSCAN
dbscan = DBSCAN(eps=epsilon, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
# Visualize the results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Cluster')
plt.title('DBSCAN Clustering - Original Data')
plt.xlabel('Standardized Balance')
plt.ylabel('Standardized Flight Miles in 12 Months')
plt.show()
```

```python
# Print number of clusters and noise points
n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
n_noise = list(clusters).count(-1)
print(f'Number of clusters: {n_clusters}')
print(f'Number of noise points: {n_noise}')
```

```python
# Calculate Silhouette Score for DBSCAN
silhouette_dbscan = silhouette_score(X, clusters)
print("Silhouette Score for DBSCAN Clustering:", silhouette_dbscan)
```

# **HEIRARCHICAL CLUSTERING**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
# Using Ward linkage
linkage_matrix = linkage(final_data, method='ward')
```

**DENDOGRAM**

```python
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Calculate the largest vertical distances
def find_largest_vertical_distances(linkage_matrix, num_distances=5): # Pass linkage_matrix as an argument
    distances = linkage_matrix[:, 2]  # Extract the distances between merged clusters
    sorted_distances = np.sort(distances)[::-1]  # Sort in descending order
    largest_distances = sorted_distances[:num_distances]  # Get the largest distances
    return largest_distances

# Assuming 'linkage_matrix' is already calculated from previous steps
largest_distances = find_largest_vertical_distances(linkage_matrix)
print("Largest vertical distances (inconsistencies):", largest_distances)
```

```python
# Cut the dendrogram to get clusters
from scipy.cluster.hierarchy import fcluster
max_d = 152  # You can adjust this value to get a different number of clusters
clusters = fcluster(linkage_matrix, max_d, criterion='distance')
n_clusters = len(np.unique(clusters))
print(f"Number of clusters: {n_clusters}")
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(final_data['Balance'], final_data['Flight_miles_12mo'], c=clusters, cmap='viridis')
plt.colorbar(scatter)
plt.title('Hierarchical Clustering Result')
plt.xlabel('Standardized Balance')
plt.ylabel('Standardized Flight_miles_12mo')
plt.show()
```

```python
# Calculate Silhouette Score for Hierarchical Clustering
silhouette_hierarchical = silhouette_score(final_data, clusters)
print("Silhouette Score of Hierarchical Clustering:", silhouette_hierarchical)
```
